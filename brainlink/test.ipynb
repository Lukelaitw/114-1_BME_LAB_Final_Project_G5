```python
import matplotlib.pyplot as plt
import numpy as np


# 如果上面的路徑不存在，可以使用專案中的文件
# biopac_data = "./Classifier/bci_dataset_113-2/S01/1.txt"

# 讀取數據

    # 如果找不到文件，使用專案中的示例文件
biopac_data = "./Classifier/bci_dataset_113-2/S01/2.txt"
data = np.loadtxt(biopac_data)
print(f"使用示例文件：{biopac_data}")
print(f"成功讀取數據，共 {len(data)} 個數據點")

# 創建時間軸（假設採樣率，可以根據實際情況調整）
# 如果知道採樣率，可以用：time = np.arange(len(data)) / sampling_rate
time = np.arange(len(data))

# 繪製圖形
plt.figure(figsize=(12, 6))
plt.plot(time, data, linewidth=0.5, alpha=0.7)
plt.xlabel('Time Points', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title('Visualization-biopac-rawdata', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# 統計信息
min_val = np.min(data)
max_val = np.max(data)
mean_val = np.mean(data)
std_val = np.std(data)
stats_text = (f"Min: {min_val:.2f}\n"
              f"Max: {max_val:.2f}\n"
              f"Mean: {mean_val:.2f}\n"
              f"Std: {std_val:.2f}")

# 將統計信息顯示在圖上
plt.gca().text(
    0.98, 0.98, stats_text,
    fontsize=12,
    color='black',
    verticalalignment='top',
    horizontalalignment='right',
    bbox=dict(facecolor='white', alpha=0.8)
    , transform=plt.gca().transAxes
)

plt.tight_layout()
plt.show()

# 顯示統計信息
print(f"\n數據統計：")
print(f"最小值: {np.min(data):.4f}")
print(f"最大值: {np.max(data):.4f}")
print(f"平均值: {np.mean(data):.4f}")

plt.gcf().text(0.75, 0.80, stats_text, fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
print(f"標準差: {np.std(data):.4f}")
```
